# Naive Bayes

### Problem Statement

    - Preparing classifiction model for Salary Data

```python
import warnings
warnings.filterwarnings('ignore')
```

### Loaing Train Data

```python
import pandas as pd
import numpy as np

train = pd.read_csv("~/desktop/Digi 360/Module 22/Datasets-6/SalaryData_Train.csv", encoding = 'latin1')
train.head()
```

```python
train.Salary.value_counts()
```

```python
# mapping salary to 0 and 1
train['Salary'] = train['Salary'].apply(lambda x: 0 if x==' <=50K' else 1)
train.head()
```

```python
train.Salary.value_counts()
```

### Encoding the categorical varibales in Train data

```python
dummies = pd.get_dummies(train[['workclass','education','maritalstatus','occupation','relationship','race', 'sex','native']],drop_first=True)
dummies.head()
```

```python
#Let's concat dummies dataframe with original dataframe and drop the original columns

train = pd.concat([train, dummies], axis = 1)
train.head()
```

```python
# Let's drop the original columns since they are already converted to numeric values.
train.drop(['workclass','education','maritalstatus','occupation','relationship','race', 'sex','native'], axis = 1, inplace = True)
train.head()
```

```python
X_train = train.drop(['Salary'], axis = 1)
X_train.head()
```

```python
y_train = train.Salary
y_train.head()
```

```python
y_train.value_counts()
```

### Loading Test Data

```python
test = pd.read_csv("~/desktop/Digi 360/Module 22/Datasets-6/SalaryData_Test.csv", encoding = 'latin1')
test.head()
```

```python
test.Salary.value_counts()
```

```python
# mapping salary to 0 and 1
test['Salary'] = test['Salary'].apply(lambda x: 0 if x==' <=50K' else 1)
test.head()
```

```python
test.Salary.value_counts()
```

### Encoding the categorical varibales in Test data

```python
dummies = pd.get_dummies(test[['workclass','education','maritalstatus','occupation','relationship','race', 'sex','native']],drop_first=True)
dummies.head()
```

```python
#Let's concat dummies dataframe with original dataframe and drop the original columns

test = pd.concat([test, dummies], axis = 1)
test.head()
```

```python
# Let's drop the original columns since they are already converted to numeric values.
test.drop(['workclass','education','maritalstatus','occupation','relationship','race', 'sex','native'], axis = 1, inplace = True)
test.head()
```

```python
X_test = test.drop(['Salary'], axis = 1)
X_test.head()
```

```python
y_test = test.Salary
y_test.head()
```

```python
y_test.value_counts()
```

### Scaling training data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
#Let's apply MinMaxScaler to the columns except the dummy varibales
num_vars = ['age', 'educationno', 'capitalgain', 'capitalloss','hoursperweek']

train[num_vars] = scaler.fit_transform(train[num_vars])
train.describe()
```

### Scaling Test Data

```python
test[num_vars] = scaler.fit_transform(test[num_vars])
test.describe()
```

### Building model

```python
#Fitting classifier to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train , y_train)
```

```python
# Predicting on test set

y_test_pred = classifier.predict(X_test)
```

```python
y_test_pred
```

```python
y_test.value_counts()
```

### Model Evaluation

```python
# printing the overall accuracy
from sklearn import metrics
metrics.accuracy_score(y_test, y_test_pred)
```

```python
metrics.confusion_matrix(y_test, y_test_pred)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_test_pred)
confusion_matrix
```

```python
confusion = metrics.confusion_matrix(y_test, y_test_pred)
print(confusion)
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]
TP = confusion[1, 1]
```

```python
sensitivity = TP / float(FN + TP)
print("sensitivity",sensitivity)
```

```python
specificity = TN / float(TN + FP)
print("specificity",specificity)
```

```python
precision = TP / float(TP + FP)
print("precision",precision)
print(metrics.precision_score(y_test, y_test_pred))
```

```python
print("precision",precision)
print("PRECISION SCORE :",metrics.precision_score(y_test, y_test_pred))
print("RECALL SCORE :", metrics.recall_score(y_test, y_test_pred))
print("F1 SCORE :",metrics.f1_score(y_test, y_test_pred))
```

### Conclusion
